```python
import pandas as pd
import numpy as np

player_match_events = pd.read_csv(
    r"C:\Users\Jesus\Desktop\Formación\Data Analytics\Proyecto final\player_match_events.csv"
)
```

```python
players["player_clean"] = (
    players["name"]
    .str.lower()
    .str.strip()
)

player_match_events["player_clean"] = (
    player_match_events["player"]
    .str.lower()
    .str.strip()
)
```

```python
intersection = set(player_match_events["player_clean"]) & set(players["player_clean"])

len(intersection)
```

```python
appearances.columns
```

```python
appearances = appearances.rename(columns={
    "game_id": "id_odsp"
})

appearances.columns
```

```python
player_match_events["id_odsp"] = player_match_events["id_odsp"].astype(str)
appearances["id_odsp"] = appearances["id_odsp"].astype(str)

```

```python
player_match_events["id_odsp"].dtype
appearances["id_odsp"].dtype
```

```python
player_match_events = player_match_events.merge(
    players[["player_id", "player_clean", "name"]],
    on="player_clean",
    how="left"
)

player_match_events["player_id"].isna().mean()

```

```python
appearances_player = appearances.groupby("player_id").agg({
    "goals": "sum",
    "assists": "sum",
    "minutes_played": "sum"
}).reset_index()
```

```python
events_final = player_match_events[
    [
        "id_odsp",
        "player_id",
        "player",
        "event_team",
        "opponent",
        "foul",
        "pass",
        "shot",
        "yellow_card",
        "second_yellow",
        "red_card",
        "total_cards"
    ]
]

events_final.to_csv(
    "C:\\Users\\Jesus\\Desktop\\Formación\\Data Analytics\\Proyecto final\\events.csv",
    index=False
)
```
