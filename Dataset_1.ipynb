```python
event_type_map = {
    1: "shot",
    2: "corner",
    3: "foul",
    4: "yellow_card",
    5: "second_yellow",
    6: "red_card",
    7: "substitution",
    8: "pass",
    9: "free_kick",
    10: "offside",
    11: "handball"
}

events["event_type_name"] = events["event_type"].map(event_type_map)

events["event_type_name"].value_counts()
```

```python
key_events = events[
    events["event_type"].isin([1, 3, 4, 5, 6, 8])
].copy()

key_events["event_type_name"].value_counts()

```

```python
player_event_counts = (
    key_events
    .groupby(["player", "event_type_name"])
    .size()
    .unstack(fill_value=0)
    .reset_index()
)

player_event_counts.head()
```

```python
player_event_counts["total_cards"] = (
    player_event_counts["yellow_card"] +
    player_event_counts["second_yellow"] +
    player_event_counts["red_card"]
)

player_event_counts["cards_per_foul"] = (
    player_event_counts["total_cards"] /
    player_event_counts["foul"]
)

player_event_counts.head()
```

```python
player_event_counts["cards_per_foul"] = (
    player_event_counts["cards_per_foul"]
    .replace([np.inf, -np.inf], np.nan)
)
```

```python
player_match_events = (
    key_events
    .groupby(["id_odsp", "player", "event_team", "opponent", "event_type_name"])
    .size()
    .unstack(fill_value=0)
    .reset_index()
)

player_match_events.head()
```

```python
player_match_events["total_cards"] = (
    player_match_events.get("yellow_card", 0) +
    player_match_events.get("second_yellow", 0) +
    player_match_events.get("red_card", 0)
)

```

```python
player_match_events = player_match_events.fillna(0)


player_match_events.to_csv(
    "C:\\Users\\Jesus\\Desktop\\Formación\\Data Analytics\\Proyecto final\\player_match_events.csv",
    index=False
)


```
